# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
# Program to calculate the value of V for given formula
import math

print("r \t V")

for r in range(2, 21, 2):  # even numbers from 2 to 20
    V = (2/3) * math.pi * (r ** 3)  # given formula volume of hemisphere
    print(r,'\t',round(V,3))

```

```python
 #Python program for simple calculator

import math

# Function to evaluate expressions with parentheses
def evaluate_expression(expression):
    try:
        # Using eval in a safe manner
        allowed_names = {'math': math, 'sqrt': math.sqrt, 'pow': math.pow, 'mod': lambda x, y: x % y}
        result = eval(expression, {"__builtins__": None}, allowed_names)
        return result
    except ZeroDivisionError:
        return "Error: Division by zero is not allowed."
    except (SyntaxError, NameError, TypeError):
        return "Error: Invalid input expression."

# Function to add two numbers
def add(num1, num2):
    return num1 + num2

# Function to subtract two numbers
def subtract(num1, num2):
    return num1 - num2

# Function to multiply two numbers
def multiply(num1, num2):
    return num1 * num2

# Function to divide two numbers
def divide(num1, num2):
    if num2 == 0:
        return "Error: Division by zero is not allowed."
    return num1 / num2

# Function to calculate power
def power(num1, num2):
    return math.pow(num1, num2)

# Function to calculate square root
def square_root(num):
    return math.sqrt(num)

# Function to calculate modulus
def modulus(num1, num2):
    return num1 % num2

#Selecting the operations
while True:
    print("Please select operation -\n"
          "1. Add\n"
          "2. Subtract\n"
          "3. Multiply\n"
          "4. Divide\n"
          "5. Power\n"
          "6. Square Root\n"
          "7. Modulus\n"
          "8. Evaluate Expression (with parentheses)\n"
          "9. Exit\n")

    select = int(input("Select operation (1-9): "))

    if select == 9:
        print("Exiting calculator. Goodbye!")
        break

    if select in [1, 2, 3, 4, 5, 7]:
        number_1 = float(input("Enter first number: "))
        number_2 = float(input("Enter second number: "))

        if select == 1:
            print(number_1, "+", number_2, "=", add(number_1, number_2))
        elif select == 2:
            print(number_1, "-", number_2, "=", subtract(number_1, number_2))
        elif select == 3:
            print(number_1, "*", number_2, "=", multiply(number_1, number_2))
        elif select == 4:
            print(number_1, "/", number_2, "=", divide(number_1, number_2))
        elif select == 5:
            print(number_1, "**", number_2, "=", power(number_1, number_2))
        elif select == 7:
            print(number_1, "%", number_2, "=", modulus(number_1, number_2))
    elif select == 6:
        number = float(input("Enter number: "))
        print("sqrt(", number, ") =", square_root(number))
    elif select == 8:
        expression = input("Enter expression: ")
        print("Result:", evaluate_expression(expression))
    else:
        print("Invalid input")

```

# Question 6 (20 points).
Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files. Finally maintaine 50 products of different categories in inventory.

```python
import csv
import os
import random

# Constants
CSV_FILE = "inventory.csv"
RESTOCK_THRESHOLD = 5  # If stock < 5, restock automatically
RESTOCK_AMOUNT = 20  # Amount to restock
CATEGORIES = ["Electronics", "Clothing", "Groceries", "Home", "Toys"]

# Random Product Names
PRODUCT_NAMES = {
    "Electronics": ["Laptop", "Smartphone", "Tablet", "Headphones", "Monitor"],
    "Clothing": ["T-Shirt", "Jeans", "Jacket", "Sneakers", "Cap"],
    "Groceries": ["Milk", "Bread", "Eggs", "Rice", "Cereal"],
    "Home": ["Sofa", "Table", "Chair", "Lamp", "Curtains"],
    "Toys": ["Action Figure", "Board Game", "Puzzle", "Doll", "RC Car"]
}

# Generate Random Products
def generate_random_products(num_products=50):

    '''
    Creates a Products list and generates the random data for the inventory using random module

    Args: we need to give the number of records to be generated as an argument
    '''
    products = []
    for i in range(num_products):
        category = random.choice(CATEGORIES)
        name = random.choice(PRODUCT_NAMES[category])
        product_id = random.randint(1000, 9999)
        stock = random.randint(1, 30)
        price = round(random.uniform(5, 1000), 2)  # Prices between $5 and $1000
        products.append([product_id, name, category, stock, price])
    return products

# Initialize CSV with random data
def initialize_inventory():

    '''
    Checks if the inventory CSV file exists or not. If the CSV file doesn't exists then it calls the function
    generate_random_products() to generate the random data.

    The random generated data will be writed in to the CSV file. Thus, this function initializes the inventory file.
    '''
    if not os.path.exists(CSV_FILE):
        products = generate_random_products()
        with open(CSV_FILE, "w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["ID", "Name", "Category", "Stock", "Price"])
            writer.writerows(products)
        print("Inventory initialized with randomly generated data.")

# Load inventory from CSV
def load_inventory():

    '''
    This loads the CSV data in to a list of dictionaries to perform other operations.

    If any exception occurs while reading the file like file is corrupted or unreadable file
    then it automatically created a new file calling initialize_inventory()
    '''
    try:
        with open(CSV_FILE, "r") as file:
            reader = csv.DictReader(file)
            return list(reader)
    except Exception:
        print("Error reading inventory file. Creating a new one.")
        initialize_inventory()
        return load_inventory()

# Save inventory back to CSV
def save_inventory(inventory):

    '''
    Once all the changes like modifying, adding or deleting any product information,
    these changes will be tracked in list of dictionaries that was created
    So, to reflec those changes in the origial CSV file this fuction overwrites the data in to CSV file.
    '''

    with open(CSV_FILE, "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["ID", "Name", "Category", "Stock", "Price"])
        for item in inventory:
            writer.writerow([item["ID"], item["Name"], item["Category"], item["Stock"], item["Price"]])

# Add new product
def add_product():

    '''
    If we want to add any new product to our inventory then we can do by using this function.
    it takes runtme inputs for the product details and adds the product in to inventory.

    Raises:
    ValueError: when invalid input is given by user like entering string literal for numerical values.
    '''

    try:
        inventory = load_inventory()
        new_id = int(input("Enter Product ID: "))
        name = input("Enter Product Name: ")
        category = input(f"Enter Category {CATEGORIES}: ")
        stock = int(input("Enter Stock Quantity: "))
        price = float(input("Enter Price: "))

        if category not in CATEGORIES:
            print("Invalid category. Product not added.")
            return

        inventory.append({"ID": new_id, "Name": name, "Category": category, "Stock": stock, "Price": price})
        save_inventory(inventory)
        print(f"Product '{name}' added successfully!")
    except ValueError:
        print("Invalid input! Please enter correct values.")


# Delete product
def delete_product():

    '''
    This function takes the produc id that has to be deleted durig runtime from user and removes the product from inventory.

    If the product id given id not found then it just prints "Product ID not found"

    '''

    inventory = load_inventory()
    product_id = input("Enter Product ID to delete: ")

    new_inventory = [item for item in inventory if item["ID"] != product_id]

    if len(new_inventory) == len(inventory):
        print("Product ID not found.")
    else:
        save_inventory(new_inventory)
        print("Product deleted successfully.")

# Modify product
def modify_product():

    '''
    This is used to modify any details of the product and save the inventory in to a CSV file
    '''
    inventory = load_inventory()
    product_id = input("Enter Product ID to modify: ")

    for item in inventory:
        if item["ID"] == product_id:
            item["Name"] = input("Enter new Name: ") or item["Name"]
            item["Category"] = input("Enter new Category: ") or item["Category"]
            item["Stock"] = input("Enter new Stock: ") or item["Stock"]
            item["Price"] = input("Enter new Price: ") or item["Price"]
            save_inventory(inventory)
            print("Product modified successfully.")
            return

    print("Product ID not found.")

# Search product
def search_product():

    '''This fundtion is used to search for any products. The search should be performed irrespective of alphabetical case.
    So, the input as well as the names in file are converted in to lower case while comparision.
    '''

    inventory = load_inventory()
    search_term = input("Enter product name or category: ").lower()
    results = [item for item in inventory if search_term in item["Name"].lower() or search_term in item["Category"].lower()]

    if results:
        print("Search Results:")
        for item in results:
            print(item)
    else:
        print("No matching products found.")

# Restock items automatically
def auto_restock():
    '''
    It checks if any of the product in the inventory is below threshold level.
    If yes then that product will be restocked automatically
    '''
    inventory = load_inventory()
    restocked_items = []

    for item in inventory:
        if int(item["Stock"]) < RESTOCK_THRESHOLD:
            item["Stock"] = str(int(item["Stock"]) + RESTOCK_AMOUNT)
            restocked_items.append(item["Name"])

    save_inventory(inventory)

    if restocked_items:
        print("Restocked:", ", ".join(restocked_items))
    else:
        print("No restocking needed.")

# Generate report
def generate_report():

    '''
    This generated the total and categorical asset report for the inventory.
    it products the quantity and price to calculates the total stock value we have.
    '''

    inventory = load_inventory()
    category_values = {}

    total_value = 0
    for item in inventory:
        value = int(item["Stock"]) * float(item["Price"])
        total_value += value
        category_values[item["Category"]] = category_values.get(item["Category"], 0) + value

    print("\n=== Inventory Report ===")
    print(f"Total Inventory Value: ${total_value:.2f}")
    print("Category Breakdown:")
    for category, value in category_values.items():
        print(f"{category}: ${value:.2f}")

# Main menu
def main():
    '''
    In the main function, a sequnce of numbers will  be shown to select ay operation
    based on the number selection, the operation will be done to the inventory
    '''
    initialize_inventory()

    while True:
        print("\nInventory Management System")
        print("1. Add Product")
        print("2. Delete Product")
        print("3. Modify Product")
        print("4. Search Product")
        print("5. Restock Items")
        print("6. Generate Report")
        print("7. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_product()
        elif choice == "2":
            delete_product()
        elif choice == "3":
            modify_product()
        elif choice == "4":
            search_product()
        elif choice == "5":
            auto_restock()
        elif choice == "6":
            generate_report()
        elif choice == "7":
            print("Exiting... Goodbye!")
            break
        else:
            print("Invalid choice. Try again.")

# Run the program
if __name__ == "__main__":
    main()
```

```python

import csv
import os
import shutil

# Constants
INVENTORY_FILE = "inventory.csv"  # File to store inventory data
BACKUP_FILE = "inventory_backup.csv"  # Backup file in case of corruption
CATEGORIES = ["Electronics", "Clothing", "Home & Kitchen", "Books", "Toys"]  # Valid product categories
MIN_STOCK_LEVEL = 10  # Minimum stock level for restocking
RESTOCK_QUANTITY = 20  # Quantity to add during restocking
MAX_PRODUCTS = 50  # Maximum number of products in inventory

# Ensure the inventory file exists and is valid
def initialize_inventory():
    """
    Initialize the inventory CSV file. If the file is missing or corrupted,
    create a new one with the correct headers.
    """
    if not os.path.exists(INVENTORY_FILE):
        # Create a new inventory file with headers
        with open(INVENTORY_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["ProductID", "Name", "Category", "Quantity", "Price"])
    else:
        # Check if the file is corrupted
        try:
            with open(INVENTORY_FILE, mode="r") as file:
                reader = csv.reader(file)
                headers = next(reader)
                if headers != ["ProductID", "Name", "Category", "Quantity", "Price"]:
                    raise ValueError("Invalid headers in CSV file.")
        except Exception as e:
            # Handle corrupted file
            print(f"Corrupted CSV file detected: {e}")
            shutil.copy(INVENTORY_FILE, BACKUP_FILE)  # Create a backup
            print(f"Backup created: {BACKUP_FILE}")
            # Initialize a new inventory file
            with open(INVENTORY_FILE, mode="w", newline="") as file:
                writer = csv.writer(file)
                writer.writerow(["ProductID", "Name", "Category", "Quantity", "Price"])
            print("New inventory file initialized.")

def load_inventory():
    """
    Load inventory data from the CSV file.
    Returns a list of dictionaries representing products.
    """
    try:
        with open(INVENTORY_FILE, mode="r") as file:
            reader = csv.DictReader(file)
            return list(reader)
    except Exception as e:
        print(f"Error loading inventory: {e}")
        return []

def save_inventory(inventory):
    """
    Save inventory data to the CSV file.
    """
    try:
        with open(INVENTORY_FILE, mode="w", newline="") as file:
            writer = csv.DictWriter(file, fieldnames=["ProductID", "Name", "Category", "Quantity", "Price"])
            writer.writeheader()
            writer.writerows(inventory)
    except Exception as e:
        print(f"Error saving inventory: {e}")

def add_product(product_id, name, category, quantity, price):
    """
    Add a new product to the inventory.
    Validates inputs and ensures no duplicate ProductID.
    """
    inventory = load_inventory()
    # Check if the maximum number of products is reached
    if len(inventory) >= MAX_PRODUCTS:
        print(f"Cannot add more than {MAX_PRODUCTS} products.")
        return
    # Check for duplicate ProductID
    if any(product["ProductID"] == product_id for product in inventory):
        print("Product ID already exists.")
        return
    # Validate category
    if category not in CATEGORIES:
        print(f"Invalid category. Valid categories are: {', '.join(CATEGORIES)}")
        return
    # Validate quantity and price
    if int(quantity) < 0 or float(price) < 0:
        print("Quantity and Price must be non-negative.")
        return
    # Add the product to the inventory
    inventory.append({
        "ProductID": product_id,
        "Name": name,
        "Category": category,
        "Quantity": quantity,
        "Price": price
    })
    save_inventory(inventory)
    print("Product added successfully.")

def delete_product(product_id):
    """
    Delete a product from the inventory by ProductID.
    """
    inventory = load_inventory()
    # Filter out the product to delete
    inventory = [product for product in inventory if product["ProductID"] != product_id]
    save_inventory(inventory)
    print("Product deleted successfully.")

def update_product(product_id, name=None, category=None, quantity=None, price=None):
    """
    Update an existing product in the inventory.
    Allows partial updates (only specified fields are updated).
    """
    inventory = load_inventory()
    for product in inventory:
        if product["ProductID"] == product_id:
            # Update name if provided
            if name:
                product["Name"] = name
            # Update category if provided
            if category:
                if category not in CATEGORIES:
                    print(f"Invalid category. Valid categories are: {', '.join(CATEGORIES)}")
                    return
                product["Category"] = category
            # Update quantity if provided
            if quantity:
                if int(quantity) < 0:
                    print("Quantity must be non-negative.")
                    return
                product["Quantity"] = quantity
            # Update price if provided
            if price:
                if float(price) < 0:
                    print("Price must be non-negative.")
                    return
                product["Price"] = price
            save_inventory(inventory)
            print("Product updated successfully.")
            return
    print("Product not found.")

def search_product(search_term):
    """
    Search for a product by name or ProductID.
    Displays matching products.
    """
    inventory = load_inventory()
    # Find products matching the search term
    results = [product for product in inventory if search_term.lower() in product["Name"].lower() or search_term == product["ProductID"]]
    if results:
        for product in results:
            print(product)
    else:
        print("No matching products found.")

def restock_inventory():
    """
    Automatically restock products with quantities below the minimum stock level.
    """
    inventory = load_inventory()
    for product in inventory:
        if int(product["Quantity"]) < MIN_STOCK_LEVEL:
            # Add restock quantity to the product
            product["Quantity"] = int(product["Quantity"]) + RESTOCK_QUANTITY
    save_inventory(inventory)
    print("Restocking completed.")

def generate_report():
    """
    Generate a report showing the total inventory value and a breakdown by category.
    """
    inventory = load_inventory()
    total_value = 0
    # Initialize category breakdown dictionary
    category_breakdown = {category: {"count": 0, "value": 0} for category in CATEGORIES}

    # Calculate total value and category breakdown
    for product in inventory:
        value = int(product["Quantity"]) * float(product["Price"])
        total_value += value
        category_breakdown[product["Category"]]["count"] += 1
        category_breakdown[product["Category"]]["value"] += value

    # Display the report
    print("\n--- Inventory Report ---")
    print(f"Total Inventory Value: ${total_value:.2f}")
    print("\nCategory Breakdown:")
    for category, data in category_breakdown.items():
        print(f"{category}: {data['count']} items, Total Value: ${data['value']:.2f}")

def main():
    """
    Main function to manage the inventory system.
    Provides a menu for user interaction.
    """
    initialize_inventory()
    while True:
        print("\n--- Inventory Management System ---")
        print("1. Add Product")
        print("2. Delete Product")
        print("3. Update Product")
        print("4. Search Product")
        print("5. Restock Inventory")
        print("6. Generate Report")
        print("7. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            # Add a new product
            product_id = input("Enter Product ID: ")
            name = input("Enter Product Name: ")
            category = input("Enter Category: ")
            quantity = input("Enter Quantity: ")
            price = input("Enter Price: ")
            try:
                quantity = int(quantity)
                price = float(price)
                add_product(product_id, name, category, quantity, price)
            except ValueError:
                print("Invalid input. Quantity must be an integer, and Price must be a number.")

        elif choice == "2":
            # Delete a product
            product_id = input("Enter Product ID to delete: ")
            delete_product(product_id)

        elif choice == "3":
            # Update a product
            product_id = input("Enter Product ID to update: ")
            name = input("Enter new Name (leave blank to skip): ")
            category = input("Enter new Category (leave blank to skip): ")
            quantity = input("Enter new Quantity (leave blank to skip): ")
            price = input("Enter new Price (leave blank to skip): ")
            try:
                quantity = int(quantity) if quantity else None
                price = float(price) if price else None
                update_product(product_id, name or None, category or None, quantity, price)
            except ValueError:
                print("Invalid input. Quantity must be an integer, and Price must be a number.")

        elif choice == "4":
            # Search for a product
            search_term = input("Enter Product Name or ID to search: ")
            search_product(search_term)

        elif choice == "5":
            # Restock inventory
            restock_inventory()

        elif choice == "6":
            # Generate inventory report
            generate_report()

        elif choice == "7":
            # Exit the program
            print("Exiting the system.")
            break

        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

```python
import csv
import os
from datetime import datetime, timedelta

# Constants
FLIGHTS_FILE = "flights.csv"
BOOKINGS_FILE = "bookings.csv"
MAX_SEATS = 60  # Maximum seats per flight
MAX_FLIGHTS = 50  # Maximum number of flights in the system

# Ensure CSV files exist and are initialized
def initialize_files():
    """
    Initialize the CSV files for flights and bookings if they don't exist.
    Also, generate example data if the files are empty.
    """
    if not os.path.exists(FLIGHTS_FILE):
        with open(FLIGHTS_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["FlightID", "Origin", "Destination", "DepartureTime", "SeatsAvailable", "Status"])
        # Generate example flight data
        generate_example_flights()

    if not os.path.exists(BOOKINGS_FILE):
        with open(BOOKINGS_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["BookingID", "FlightID", "PassengerName", "SeatNumber", "SeatType"])
        # Generate example booking data
        generate_example_bookings()

def generate_example_flights():
    """
    Generate example flight data and save it to the CSV file.
    """
    flights = []
    current_time = datetime.now()
    for i in range(1, 11):  # Generate 10 example flights
        flight_id = f"FL{i:03d}"  # Flight IDs like FL001, FL002, etc.
        origin = "New York" if i % 2 == 0 else "Los Angeles"
        destination = "Los Angeles" if i % 2 == 0 else "New York"
        departure_time = (current_time + timedelta(hours=i)).strftime("%Y-%m-%d %H:%M")
        seats_available = MAX_SEATS - (i * 5)  # Simulate some bookings
        status = "On-Time" if i % 2 == 0 else "Delayed"
        flights.append({
            "FlightID": flight_id,
            "Origin": origin,
            "Destination": destination,
            "DepartureTime": departure_time,
            "SeatsAvailable": seats_available,
            "Status": status
        })
    save_flights(flights)

def generate_example_bookings():
    """
    Generate example booking data and save it to the CSV file.
    """
    bookings = []
    for i in range(1, 21):  # Generate 20 example bookings
        booking_id = i
        flight_id = f"FL{(i % 10) + 1:03d}"  # Distribute bookings across 10 flights
        passenger_name = f"Passenger{i}"
        seat_number = (i % MAX_SEATS) + 1
        seat_type = "Window" if i % 2 == 0 else "Aisle"
        bookings.append({
            "BookingID": booking_id,
            "FlightID": flight_id,
            "PassengerName": passenger_name,
            "SeatNumber": seat_number,
            "SeatType": seat_type
        })
    save_bookings(bookings)

def load_flights():
    """
    Load flight data from the CSV file.
    """
    try:
        with open(FLIGHTS_FILE, mode="r") as file:
            reader = csv.DictReader(file)
            return list(reader)
    except Exception as e:
        print(f"Error loading flights: {e}")
        return []

def load_bookings():
    """
    Load booking data from the CSV file.
    """
    try:
        with open(BOOKINGS_FILE, mode="r") as file:
            reader = csv.DictReader(file)
            return list(reader)
    except Exception as e:
        print(f"Error loading bookings: {e}")
        return []

def save_flights(flights):
    """
    Save flight data to the CSV file.
    """
    try:
        with open(FLIGHTS_FILE, mode="w", newline="") as file:
            writer = csv.DictWriter(file, fieldnames=["FlightID", "Origin", "Destination", "DepartureTime", "SeatsAvailable", "Status"])
            writer.writeheader()
            writer.writerows(flights)
    except Exception as e:
        print(f"Error saving flights: {e}")

def save_bookings(bookings):
    """
    Save booking data to the CSV file.
    """
    try:
        with open(BOOKINGS_FILE, mode="w", newline="") as file:
            writer = csv.DictWriter(file, fieldnames=["BookingID", "FlightID", "PassengerName", "SeatNumber", "SeatType"])
            writer.writeheader()
            writer.writerows(bookings)
    except Exception as e:
        print(f"Error saving bookings: {e}")

def add_flight(flight_id, origin, destination, departure_time, status="On-Time"):
    """
    Add a new flight to the system.
    """
    flights = load_flights()
    if len(flights) >= MAX_FLIGHTS:
        print(f"Cannot add more than {MAX_FLIGHTS} flights.")
        return
    if any(flight["FlightID"] == flight_id for flight in flights):
        print("Flight ID already exists.")
        return
    flights.append({
        "FlightID": flight_id,
        "Origin": origin,
        "Destination": destination,
        "DepartureTime": departure_time,
        "SeatsAvailable": MAX_SEATS,
        "Status": status
    })
    save_flights(flights)
    print("Flight added successfully.")

def find_flights(origin, destination):
    """
    Find available flights based on origin and destination.
    """
    flights = load_flights()
    results = [flight for flight in flights if flight["Origin"] == origin and flight["Destination"] == destination and int(flight["SeatsAvailable"]) > 0]
    if results:
        print("\nAvailable Flights:")
        for flight in results:
            print(f"Flight ID: {flight['FlightID']}, Departure: {flight['DepartureTime']}, Seats Available: {flight['SeatsAvailable']}, Status: {flight['Status']}")
    else:
        print("No available flights found.")

def reserve_seat(flight_id, passenger_name, seat_type="Regular"):
    """
    Reserve a seat on a flight.
    """
    flights = load_flights()
    bookings = load_bookings()
    flight = next((f for f in flights if f["FlightID"] == flight_id), None)
    if not flight:
        print("Flight not found.")
        return
    if int(flight["SeatsAvailable"]) <= 0:
        print("No seats available on this flight.")
        return
    # Assign a seat number
    seat_number = MAX_SEATS - int(flight["SeatsAvailable"]) + 1
    # Update flight seats
    flight["SeatsAvailable"] = str(int(flight["SeatsAvailable"]) - 1)
    # Add booking
    booking_id = len(bookings) + 1
    bookings.append({
        "BookingID": booking_id,
        "FlightID": flight_id,
        "PassengerName": passenger_name,
        "SeatNumber": seat_number,
        "SeatType": seat_type
    })
    save_flights(flights)
    save_bookings(bookings)
    print(f"Seat {seat_number} ({seat_type}) reserved for {passenger_name} on Flight {flight_id}.")

def cancel_booking(booking_id):
    """
    Cancel a booking and release the seat.
    """
    bookings = load_bookings()
    booking = next((b for b in bookings if b["BookingID"] == booking_id), None)
    if not booking:
        print("Booking not found.")
        return
    # Release the seat
    flights = load_flights()
    flight = next((f for f in flights if f["FlightID"] == booking["FlightID"]), None)
    if flight:
        flight["SeatsAvailable"] = str(int(flight["SeatsAvailable"]) + 1)
    # Remove the booking
    bookings = [b for b in bookings if b["BookingID"] != booking_id]
    save_flights(flights)
    save_bookings(bookings)
    print(f"Booking {booking_id} canceled successfully.")

def generate_report():
    """
    Generate a performance report showing flight and booking statistics.
    """
    flights = load_flights()
    bookings = load_bookings()
    total_flights = len(flights)
    total_bookings = len(bookings)
    total_revenue = total_bookings * 100  # Assuming $100 per booking

    print("\n--- Flight Booking System Report ---")
    print(f"Total Flights: {total_flights}")
    print(f"Total Bookings: {total_bookings}")
    print(f"Total Revenue: ${total_revenue}")
    print("\nFlight Status Summary:")
    status_summary = {}
    for flight in flights:
        status = flight["Status"]
        status_summary[status] = status_summary.get(status, 0) + 1
    for status, count in status_summary.items():
        print(f"{status}: {count} flights")

def main():
    """
    Main function to manage the flight booking system.
    """
    initialize_files()
    while True:
        print("\n--- Flight Booking System ---")
        print("1. Add Flight")
        print("2. Find Flights")
        print("3. Reserve Seat")
        print("4. Cancel Booking")
        print("5. Generate Report")
        print("6. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            flight_id = input("Enter Flight ID: ")
            origin = input("Enter Origin: ")
            destination = input("Enter Destination: ")
            departure_time = input("Enter Departure Time (YYYY-MM-DD HH:MM): ")
            add_flight(flight_id, origin, destination, departure_time)

        elif choice == "2":
            origin = input("Enter Origin: ")
            destination = input("Enter Destination: ")
            find_flights(origin, destination)

        elif choice == "3":
            flight_id = input("Enter Flight ID: ")
            passenger_name = input("Enter Passenger Name: ")
            seat_type = input("Enter Seat Type (Regular/Window/Aisle): ")
            reserve_seat(flight_id, passenger_name, seat_type)

        elif choice == "4":
            booking_id = input("Enter Booking ID: ")
            cancel_booking(int(booking_id))

        elif choice == "5":
            generate_report()

        elif choice == "6":
            print("Exiting the system.")
            break

        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

```python
# Prompt 2: Avoid overbooking of the seats. Handle cases where users try to book or cancel using a non-existent flight ID. Prevent a passenger from booking multiple  seats on the same flight.Ensure that a  passenger cannot cancel a booking they never made.Handle cases where no flights exist between a given origin and destination.Prevent bookings for flights that have already departed.Ensure that empty or invalid input values do not cause system crashes.The system should still generate meaningful reports even if no reservations have been made.Ensure the system does not allow multiple rapid-fire requests to break consistency.The system should efficiently handle a large number of concurrent booking requests without slowdowns or crashes......update the above code using these edge cases so that code should execute these edge cases successfully

import sys
from datetime import datetime
from collections import defaultdict
from threading import Lock

class Flight:
    def __init__(self, flight_id, origin, destination, departure_time, total_seats):
        self.flight_id = flight_id
        self.origin = origin
        self.destination = destination
        self.departure_time = datetime.strptime(departure_time, "%Y-%m-%d %H:%M")
        self.total_seats = total_seats
        self.available_seats = total_seats
        self.reservations = {}  # {seat_number: passenger_id}
        self.status = "Scheduled"
        self.lock = Lock()  # To handle concurrent requests

    def reserve_seat(self, passenger_id, seat_number):
        with self.lock:
            if self.available_seats <= 0:
                raise ValueError("No available seats. Flight is fully booked.")
            if seat_number in self.reservations:
                raise ValueError(f"Seat {seat_number} is already reserved.")
            if passenger_id in self.reservations.values():
                raise ValueError("Passenger already has a reservation on this flight.")
            if datetime.now() > self.departure_time:
                raise ValueError("Flight has already departed. Cannot book seats.")
            self.reservations[seat_number] = passenger_id
            self.available_seats -= 1

    def release_seat(self, passenger_id, seat_number):
        with self.lock:
            if seat_number not in self.reservations:
                raise ValueError(f"Seat {seat_number} is not reserved.")
            if self.reservations[seat_number] != passenger_id:
                raise ValueError("Passenger did not reserve this seat.")
            del self.reservations[seat_number]
            self.available_seats += 1

    def update_status(self, status):
        with self.lock:
            self.status = status

    def __str__(self):
        return (f"Flight {self.flight_id}: {self.origin} -> {self.destination}, "
                f"Departs at {self.departure_time}, Status: {self.status}, "
                f"Available Seats: {self.available_seats}/{self.total_seats}")


class FlightBookingSystem:
    def __init__(self):
        self.flights = {}  # {flight_id: Flight}
        self.loyalty_points = defaultdict(int)
        self.system_lock = Lock()  # To handle concurrent operations on the system

    def add_flight(self, flight_id, origin, destination, departure_time, total_seats):
        with self.system_lock:
            if flight_id in self.flights:
                raise ValueError("Flight ID already exists.")
            if not flight_id or not origin or not destination or not departure_time or total_seats <= 0:
                raise ValueError("Invalid input values for flight.")
            self.flights[flight_id] = Flight(flight_id, origin, destination, departure_time, total_seats)

    def find_flights(self, origin, destination):
        with self.system_lock:
            if not origin or not destination:
                raise ValueError("Origin and destination cannot be empty.")
            matching_flights = [flight for flight in self.flights.values() if flight.origin == origin and flight.destination == destination]
            if not matching_flights:
                raise ValueError("No flights found for the given origin and destination.")
            return matching_flights

    def reserve_seat(self, flight_id, passenger_id, seat_number):
        with self.system_lock:
            if flight_id not in self.flights:
                raise ValueError("Flight not found.")
            if not passenger_id or not seat_number:
                raise ValueError("Passenger ID and seat number cannot be empty.")
            self.flights[flight_id].reserve_seat(passenger_id, seat_number)
            self.loyalty_points[passenger_id] += 10  # Award loyalty points

    def release_seat(self, flight_id, passenger_id, seat_number):
        with self.system_lock:
            if flight_id not in self.flights:
                raise ValueError("Flight not found.")
            if not passenger_id or not seat_number:
                raise ValueError("Passenger ID and seat number cannot be empty.")
            self.flights[flight_id].release_seat(passenger_id, seat_number)
            self.loyalty_points[passenger_id] -= 10  # Deduct loyalty points

    def cancel_reservation(self, flight_id, passenger_id):
        with self.system_lock:
            if flight_id not in self.flights:
                raise ValueError("Flight not found.")
            if not passenger_id:
                raise ValueError("Passenger ID cannot be empty.")
            flight = self.flights[flight_id]
            seats_to_release = [seat for seat, pid in flight.reservations.items() if pid == passenger_id]
            if not seats_to_release:
                raise ValueError("Passenger has no reservations on this flight.")
            for seat in seats_to_release:
                flight.release_seat(passenger_id, seat)
            self.loyalty_points[passenger_id] -= 10  # Deduct loyalty points

    def generate_report(self):
        with self.system_lock:
            report = {
                "total_flights": len(self.flights),
                "total_reservations": sum(len(flight.reservations) for flight in self.flights.values()),
                "total_available_seats": sum(flight.available_seats for flight in self.flights.values()),
                "flights": [str(flight) for flight in self.flights.values()]
            }
            return report

    def update_flight_status(self, flight_id, status):
        with self.system_lock:
            if flight_id not in self.flights:
                raise ValueError("Flight not found.")
            if not status:
                raise ValueError("Status cannot be empty.")
            self.flights[flight_id].update_status(status)

    def get_loyalty_points(self, passenger_id):
        with self.system_lock:
            if not passenger_id:
                raise ValueError("Passenger ID cannot be empty.")
            return self.loyalty_points.get(passenger_id, 0)


# User Interaction
def display_menu():
    print("\nFlight Booking System")
    print("1. Add Flight")
    print("2. Find Flights")
    print("3. Reserve Seat")
    print("4. Cancel Reservation")
    print("5. Generate Report")
    print("6. Update Flight Status")
    print("7. Check Loyalty Points")
    print("8. Exit")


def main():
    system = FlightBookingSystem()

    while True:
        display_menu()
        choice = input("Enter your choice: ")

        if choice == "1":
            try:
                flight_id = input("Enter Flight ID: ")
                origin = input("Enter Origin: ")
                destination = input("Enter Destination: ")
                departure_time = input("Enter Departure Time (YYYY-MM-DD HH:MM): ")
                total_seats = int(input("Enter Total Seats: "))
                system.add_flight(flight_id, origin, destination, departure_time, total_seats)
                print("Flight added successfully!")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "2":
            try:
                origin = input("Enter Origin: ")
                destination = input("Enter Destination: ")
                flights = system.find_flights(origin, destination)
                for flight in flights:
                    print(flight)
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "3":
            try:
                flight_id = input("Enter Flight ID: ")
                passenger_id = input("Enter Passenger ID: ")
                seat_number = input("Enter Seat Number: ")
                system.reserve_seat(flight_id, passenger_id, seat_number)
                print("Seat reserved successfully!")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "4":
            try:
                flight_id = input("Enter Flight ID: ")
                passenger_id = input("Enter Passenger ID: ")
                system.cancel_reservation(flight_id, passenger_id)
                print("Reservation canceled successfully!")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "5":
            report = system.generate_report()
            print("\nPerformance Report:")
            for key, value in report.items():
                print(f"{key}: {value}")

        elif choice == "6":
            try:
                flight_id = input("Enter Flight ID: ")
                status = input("Enter New Status: ")
                system.update_flight_status(flight_id, status)
                print("Flight status updated successfully!")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "7":
            try:
                passenger_id = input("Enter Passenger ID: ")
                points = system.get_loyalty_points(passenger_id)
                print(f"Loyalty Points for {passenger_id}: {points}")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "8":
            print("Exiting the system. Goodbye!")
            break

        else:
            print("Invalid choice. Please try again.")


if __name__ == "__main__":
    main()

```

```python
import sys
from datetime import datetime
from collections import defaultdict
from threading import Lock
import pytz  # For time zone handling

class Flight:
    def __init__(self, flight_id, origin, destination, departure_time, total_seats, origin_timezone, destination_timezone):
        self.flight_id = flight_id
        self.origin = origin
        self.destination = destination
        self.origin_timezone = pytz.timezone(origin_timezone)  # Time zone of origin
        self.destination_timezone = pytz.timezone(destination_timezone)  # Time zone of destination
        self.departure_time = self.origin_timezone.localize(datetime.strptime(departure_time, "%Y-%m-%d %H:%M"))  # Localized departure time
        self.total_seats = total_seats
        self.available_seats = total_seats
        self.reservations = {}  # {seat_number: passenger_id}
        self.status = "Scheduled"
        self.lock = Lock()  # To handle concurrent requests

    def reserve_seat(self, passenger_id, seat_number):
        with self.lock:
            if self.available_seats <= 0:
                raise ValueError("No available seats. Flight is fully booked.")
            if seat_number in self.reservations:
                raise ValueError(f"Seat {seat_number} is already reserved.")
            if passenger_id in self.reservations.values():
                raise ValueError("Passenger already has a reservation on this flight.")
            if datetime.now(pytz.utc) > self.departure_time.astimezone(pytz.utc):
                raise ValueError("Flight has already departed. Cannot book seats.")
            self.reservations[seat_number] = passenger_id
            self.available_seats -= 1

    def release_seat(self, passenger_id, seat_number):
        with self.lock:
            if seat_number not in self.reservations:
                raise ValueError(f"Seat {seat_number} is not reserved.")
            if self.reservations[seat_number] != passenger_id:
                raise ValueError("Passenger did not reserve this seat.")
            del self.reservations[seat_number]
            self.available_seats += 1

    def update_status(self, status):
        with self.lock:
            self.status = status

    def __str__(self):
        return (f"Flight {self.flight_id}: {self.origin} -> {self.destination}, "
                f"Departs at {self.departure_time.strftime('%Y-%m-%d %H:%M %Z')}, Status: {self.status}, "
                f"Available Seats: {self.available_seats}/{self.total_seats}")


class FlightBookingSystem:
    def __init__(self):
        self.flights = {}  # {flight_id: Flight}
        self.loyalty_points = defaultdict(int)
        self.system_lock = Lock()  # To handle concurrent operations on the system

    def add_flight(self, flight_id, origin, destination, departure_time, total_seats, origin_timezone, destination_timezone):
        with self.system_lock:
            if flight_id in self.flights:
                raise ValueError("Flight ID already exists.")
            if not flight_id or not origin or not destination or not departure_time or total_seats <= 0:
                raise ValueError("Invalid input values for flight.")
            self.flights[flight_id] = Flight(flight_id, origin, destination, departure_time, total_seats, origin_timezone, destination_timezone)

    def find_flights(self, origin, destination):
        with self.system_lock:
            if not origin or not destination:
                raise ValueError("Origin and destination cannot be empty.")
            matching_flights = [flight for flight in self.flights.values() if flight.origin == origin and flight.destination == destination]
            if not matching_flights:
                raise ValueError("No flights found for the given origin and destination.")
            return matching_flights

    def reserve_seat(self, flight_id, passenger_id, seat_number):
        with self.system_lock:
            if flight_id not in self.flights:
                raise ValueError("Flight not found.")
            if not passenger_id or not seat_number:
                raise ValueError("Passenger ID and seat number cannot be empty.")
            self.flights[flight_id].reserve_seat(passenger_id, seat_number)
            self.loyalty_points[passenger_id] += 10  # Award loyalty points

    def release_seat(self, flight_id, passenger_id, seat_number):
        with self.system_lock:
            if flight_id not in self.flights:
                raise ValueError("Flight not found.")
            if not passenger_id or not seat_number:
                raise ValueError("Passenger ID and seat number cannot be empty.")
            self.flights[flight_id].release_seat(passenger_id, seat_number)
            self.loyalty_points[passenger_id] -= 10  # Deduct loyalty points

    def cancel_reservation(self, flight_id, passenger_id):
        with self.system_lock:
            if flight_id not in self.flights:
                raise ValueError("Flight not found.")
            if not passenger_id:
                raise ValueError("Passenger ID cannot be empty.")
            flight = self.flights[flight_id]
            seats_to_release = [seat for seat, pid in flight.reservations.items() if pid == passenger_id]
            if not seats_to_release:
                raise ValueError("Passenger has no reservations on this flight.")
            for seat in seats_to_release:
                flight.release_seat(passenger_id, seat)
            self.loyalty_points[passenger_id] -= 10  # Deduct loyalty points

    def generate_report(self):
        with self.system_lock:
            report = {
                "total_flights": len(self.flights),
                "total_reservations": sum(len(flight.reservations) for flight in self.flights.values()),
                "total_available_seats": sum(flight.available_seats for flight in self.flights.values()),
                "flights": [str(flight) for flight in self.flights.values()]
            }
            return report

    def update_flight_status(self, flight_id, status):
        with self.system_lock:
            if flight_id not in self.flights:
                raise ValueError("Flight not found.")
            if not status:
                raise ValueError("Status cannot be empty.")
            self.flights[flight_id].update_status(status)

    def get_loyalty_points(self, passenger_id):
        with self.system_lock:
            if not passenger_id:
                raise ValueError("Passenger ID cannot be empty.")
            return self.loyalty_points.get(passenger_id, 0)


# User Interaction
def display_menu():
    print("\nFlight Booking System")
    print("1. Add Flight")
    print("2. Find Flights")
    print("3. Reserve Seat")
    print("4. Cancel Reservation")
    print("5. Generate Report")
    print("6. Update Flight Status")
    print("7. Check Loyalty Points")
    print("8. Exit")


def main():
    system = FlightBookingSystem()

    while True:
        display_menu()
        choice = input("Enter your choice: ")

        if choice == "1":
            try:
                flight_id = input("Enter Flight ID: ")
                origin = input("Enter Origin: ")
                destination = input("Enter Destination: ")
                departure_time = input("Enter Departure Time (YYYY-MM-DD HH:MM): ")
                total_seats = int(input("Enter Total Seats: "))
                origin_timezone = input("Enter Origin Timezone (e.g., 'America/New_York'): ")
                destination_timezone = input("Enter Destination Timezone (e.g., 'America/Los_Angeles'): ")
                system.add_flight(flight_id, origin, destination, departure_time, total_seats, origin_timezone, destination_timezone)
                print("Flight added successfully!")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "2":
            try:
                origin = input("Enter Origin: ")
                destination = input("Enter Destination: ")
                flights = system.find_flights(origin, destination)
                for flight in flights:
                    print(flight)
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "3":
            try:
                flight_id = input("Enter Flight ID: ")
                passenger_id = input("Enter Passenger ID: ")
                seat_number = input("Enter Seat Number: ")
                system.reserve_seat(flight_id, passenger_id, seat_number)
                print("Seat reserved successfully!")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "4":
            try:
                flight_id = input("Enter Flight ID: ")
                passenger_id = input("Enter Passenger ID: ")
                system.cancel_reservation(flight_id, passenger_id)
                print("Reservation canceled successfully!")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "5":
            report = system.generate_report()
            print("\nPerformance Report:")
            for key, value in report.items():
                print(f"{key}: {value}")

        elif choice == "6":
            try:
                flight_id = input("Enter Flight ID: ")
                status = input("Enter New Status: ")
                system.update_flight_status(flight_id, status)
                print("Flight status updated successfully!")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "7":
            try:
                passenger_id = input("Enter Passenger ID: ")
                points = system.get_loyalty_points(passenger_id)
                print(f"Loyalty Points for {passenger_id}: {points}")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "8":
            print("Exiting the system. Goodbye!")
            break

        else:
            print("Invalid choice. Please try again.")


if __name__ == "__main__":
    main()
```
